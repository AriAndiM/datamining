Load Dataset Iris

```python
colnames = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']
url = "https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv"
iris = pd.read_csv(url)
iris
```

Drop fitur label

```python
x = iris.drop(columns='variety')
x
```

```python
# percent_amount_of_test_data = / HUNDRED_PERCENT
# percent_amount_of_test_data = 0.3
```

Memisahkan values dan class

```python
# separate target 

# untuk mengambil values data
x=iris.iloc[:,0:4].values

# untuk mengambil classes label
y=iris.iloc[:,4].values
```

Splitting Data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

Hasil Akurasi Naive Bayes Gaussian

```python
gnb = GaussianNB()

gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)
akurasi_nb = round(metrics.accuracy_score(y_test, y_pred)*100, 2)
print('Akurasi Gaussian Naive Bayes :', akurasi_nb, '%')
```

Hasil Prediksi

```python
data_pred = pd.concat([pd.DataFrame(x_test, columns=colnames[:-1]), pd.DataFrame(y_test, columns=[colnames[-1]])], axis=1)
data_pred
```
